```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS, ImageColorGenerator
```

```python
df=pd.read_csv('/content/tratamento tweets2.csv',encoding= 'utf-8',sep=';')
```

```python
df.head(10)
```

```python
df.columns
```

```python
Tweets= df.dropna(subset=['Tweets'],axis=0)['Tweets']
All_Tweets = " ".join(s for s in Tweets)
```

```python
stopwords = set(STOPWORDS)
stopwords.update(['da','do','na','DE','ao','os','e','para','com','de','das','mas','co','ele','https','por','seu','está','mas','minha','se','eles','até','assim','já','uma','sou','eu','san','que','um','são','isso','nós','foi','não',''])
```

```python
#df['Tweets'].value_counts()
```

```python
NUVEM = WordCloud(stopwords=stopwords,background_color='black',width=1600,height=800).generate(All_Tweets)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.imshow(NUVEM, interpolation='bilinear')
ax.set_axis_off()
```
